## NumPy Task: Compute moving average
You are given daily sales figures.

1) Compute 3-day moving average (valid mode)
2) Find the maximum moving average value


```python
import numpy as np

sales = np.array([10.0, 12.0, 9.0, 15.0, 18.0, 14.0])

# TODO: compute 3-day moving average
mov_avg = None

# TODO: compute max of moving average
max_avg = None

window = 3
expected_mov = np.convolve(sales, np.ones(window) / window, mode="valid")
expected_max = expected_mov.max()

assert mov_avg is not None, "mov_avg is None."
assert max_avg is not None, "max_avg is None."
assert np.allclose(mov_avg, expected_mov), f"Expected {expected_mov}, got {mov_avg}"
assert abs(max_avg - expected_max) < 1e-12, f"Expected {expected_max}, got {max_avg}"
```

## Pandas Task: Pivot and aggregate
You are given time-tracking data.

1) Create a pivot table of total hours by employee and project
2) Return total hours per employee sorted descending


```python
import pandas as pd

time = pd.DataFrame({
    "employee": ["Ana", "Ana", "Ben", "Ben", "Cara"],
    "project": ["X", "Y", "X", "Y", "X"],
    "hours": [4, 3, 5, 2, 6],
})

# TODO: create pivot table
pivot = None

# TODO: sum hours per employee from pivot
hours_by_employee = None

expected_pivot = time.pivot_table(index="employee", columns="project", values="hours", aggfunc="sum", fill_value=0)
expected = expected_pivot.sum(axis=1).sort_values(ascending=False)

assert pivot is not None, "pivot is None."
assert hours_by_employee is not None, "hours_by_employee is None."
assert pivot.equals(expected_pivot), f"Expected {expected_pivot}, got {pivot}"
assert hours_by_employee.equals(expected), f"Expected {expected}, got {hours_by_employee}"
```
