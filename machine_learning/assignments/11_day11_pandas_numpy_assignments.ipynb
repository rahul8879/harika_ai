## Pandas Task: Group and normalize
You are given expense data.

1) Compute total spend per department
2) Normalize each department total by overall total
3) Return normalized totals sorted descending


```python
import pandas as pd

expenses = pd.DataFrame({
    "department": ["Ops", "Ops", "HR", "Eng", "Eng"],
    "amount": [200, 150, 100, 300, 250],
})

# TODO: group by department and sum
dept_total = None

# TODO: normalize by overall total
dept_share = None

expected_total = expenses.groupby("department")["amount"].sum()
expected_share = (expected_total / expected_total.sum()).sort_values(ascending=False)

assert dept_total is not None, "dept_total is None."
assert dept_share is not None, "dept_share is None."
assert dept_share.equals(expected_share), f"Expected {expected_share}, got {dept_share}"
```
